```python
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
from collections import namedtuple
```

```python
widths = np.array([.016, .008, .004, .002])
lengths = np.array([2.20,4.00,8.41,19.6])

plt.plot(widths, lengths, '+')
```

```python
invw = 1./widths
plt.plot(invw, lengths, '+')
lintuple = namedtuple('results', 'slope inter r_value p_value stderr')
results = lintuple._make(linregress(invw, lengths))
print(results)
xs = np.linspace(np.min(invw), np.max(invw), 200)
ys = results.slope*xs+results.inter
_ = plt.plot(xs, ys)
```

```python
plt.plot(widths, lengths, '+')
plt.plot(1/xs, ys)
```
